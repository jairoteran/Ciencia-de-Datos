## 4Ô∏è‚É£ Identificaci√≥n de variables num√©ricas y categ√≥ricas

```python

numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
categorical_cols = df.select_dtypes(include=["object","category"]).columns.tolist()

numeric_cols, categorical_cols
```


## 5Ô∏è‚É£ Visualizaciones del EDA  
### A) Histogramas ‚Äî Distribuci√≥n de variables num√©ricas  
Permiten observar si la distribuci√≥n es:
- Normal  
- Sesgada a la derecha/izquierda  
- Uniforme  


```python

for col in numeric_cols[:3]:  # mostramos 3 para claridad
    plt.hist(df[col], bins=20)
    plt.title(f"Histograma de {col}")
    plt.xlabel(col)
    plt.ylabel("Frecuencia")
    plt.show()
```


### B) Boxplots ‚Äî Identificaci√≥n de outliers y dispersi√≥n  
Permite observar:
- Si hay valores at√≠picos  
- D√≥nde se encuentra la mediana  
- Si la distribuci√≥n es sim√©trica  


```python

for col in numeric_cols[:3]:
    plt.boxplot(df[col].dropna(), vert=False)
    plt.title(f"Boxplot de {col}")
    plt.xlabel(col)
    plt.show()
```


### C) Gr√°fico de barras ‚Äî Variable categ√≥rica  
Se analiza la frecuencia de una categor√≠a.  


```python

cat = categorical_cols[0]
df[cat].value_counts().plot(kind="bar")
plt.title(f"Frecuencia por categor√≠a: {cat}")
plt.xlabel(cat)
plt.ylabel("Conteo")
plt.show()
```


### D) Gr√°fico de dispersi√≥n  
Permite detectar patrones o tendencias entre dos variables num√©ricas.  


```python

plt.scatter(df[numeric_cols[0]], df[numeric_cols[1]])
plt.xlabel(numeric_cols[0])
plt.ylabel(numeric_cols[1])
plt.title(f"Dispersi√≥n entre {numeric_cols[0]} y {numeric_cols[1]}")
plt.show()
```
